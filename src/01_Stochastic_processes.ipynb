# Stochastic processes

## Definition

Suppose we observe a stock price of a company at every fixed instance $t$ after some initial time $t_0$ and know $T$.

- We can interpret the observed stock values as a realization $X(t)(\omega)$ of the random variable $X(t)$ (kind of parametric)
- We look for a model which takes into account almost continuous realizations of the stock prices.

A stockastic process $X(t)$ is a collection of random variables : $$(X(t), t \in T ) = (X(t)(\omega), t \in T, \omega \in \Omega)$$. They are often used to model noise of randomness in the data.

$X(t)$ is a function of two variables:

- For a fixed instant $t$ it's a random variable: $X(t)=X(t)(\omega)\space for\space \omega\space \in\Omega$
- For a fixed random outcome $\omega$ it's a function of time: $X(t)=X(t)(\omega)\space for\space t\space \in T$

## Wiener processes (or Brownian motion)

A stochastic process $W(t)$ for $t$ in $[\,0,\infty)\,$ is called a Winer process (or a Brownian motion) if the following conditions are satisfied:

- It starts at zero : $W_0 = 0$
- It has stationary, independent increments
- For every $t>0$, $W(t)$ has a normal distribution with mean 0 and variance $t$.
- It has a.s. continuous paths with no jumps

$$dW(t) \longrightarrow W(t+\Delta t)-W(t) = \Delta W(t)$$
Thus $dW(t)$ is from normal distribution; The mean of $dW(t)=0$ then $\mathbb{E}(dW(t))=0$ and the variance is $dt$ then $dW(t)=\sqrt{dt}.Z$ where $Z$ is normally distributed with mean=0 and variance=1.

Some examples:

- Arithmetic Brownian Motion (ABM), Bachelier motion (1905), "linear function". Mostly used to model interest rates: $$dX(t)=\mu dt+\sigma dW(t)$$
- Geometric Brownian Motion (GBM), "exponential function". Mostly used to model stock behavior: $$dX(t)=\mu X(t)dt+\sigma X(t) dW(t)$$
- Ornstein-Uhlenbeck mean reverting processes (OU), "oscillation function". Mostly used to model interest rates : $$dX(t)=\kappa(\theta-X(t)) dt+\sigma dW(t)$$

Where :

- $\mu$ is the drift term
- $\sigma$ is the volatility term (dilution)
- $\kappa$ speed of mean reversion
- $\theta$ long term mean

They are known constants and $W(t)$ is a Wiener process (stochastic process).

# Discretization and implementation

Given the GBM process : $$dS(t)=\mu S(t)dt+\sigma S(t) dW(t)$$
Under the log-transformation $X(t) = log(S(t))$ we have an ABM:
$$dX(t)=(\mu-\frac{1}{2}\sigma^2)dt+ \sigma dW(t)$$
Then we discretized it and obtain the following:
$$X(t+\Delta t) = X(t)+(\mu-\frac{1}{2}\sigma^2)\Delta t+ \sigma (W(t+\Delta T)-W(t))$$
