# Martingale

A filtration $\mathcal{F}(t)$ means roughly that at some point $t$ everything happened until time $t$ is known (path...).

A stochastic process $X=\{X(t); t\geq0\}$ with respect to the filtration $\{\mathcal{F}(t)\}$, is a Martingale if:

- $X$ is adapted to the filtration $\mathcal{F}_{ t\geq0}$ : we know the value of the path before $t$.
- For all $t$, $\mathbb{E}(|X(t)|)<\infty$ : The process doesn't have exploding moment.
- For all $s$ and $t$ with $s<t$, we have $\mathbb{E}(X(t)|\mathcal{F}(s))=X(s)$ : No arbitrage

A stochastic integral process is a martingale:
Let $g \in \mathcal{L}^2$, For any $t\geq 0$: $$X(t)=\int_{0}^{t} g(s)\space \mathrm{d}W(s)\Longrightarrow dX(t)=g(t)dW(t)$$ Then the process $X=X(t); t\geq 0$ is a $\mathcal{F}(t)$-martingale. This mean whenever we have a process with no drift $\mu$ then it's a martingale.
